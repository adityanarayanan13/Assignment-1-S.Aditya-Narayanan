```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('StudentsPerformance.csv')
# Set the style for the boxplot
sns.set(style="whitegrid")

# Melt the DataFrame to have a long-form DataFrame for easier plotting
scores_melted = df.melt(value_vars=['Math Score', 'Reading Score', 'Writing Score'],
                         var_name='Subject', value_name='Score')

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Subject', y='Score', data=scores_melted, palette='pastel')

# Add title and labels
plt.title('Boxplots of Student Scores')
plt.xlabel('Subject')
plt.ylabel('Scores')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('StudentsPerformance.csv')

# Calculate the average score for each score type
math_avg = df['Math Score'].mean()
reading_avg = df['Reading Score'].mean()
writing_avg = df['Writing Score'].mean()

# Filter the data for students who scored above average in all three subjects
above_avg_students = df[
    (df['Math Score'] > math_avg) &
    (df['Reading Score'] > reading_avg) &
    (df['Writing Score'] > writing_avg)
]

# Count the number of male and female students who scored above average
gender_counts_above_avg = above_avg_students['Gender'].value_counts()

# Plot a bar graph
gender_counts_above_avg.plot(kind='bar', color=['lightblue', 'pink'])

# Add labels and title
plt.title('Male and Female Students Scoring Above Average')
plt.xlabel('Gender')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('StudentsPerformance.csv')

# Create a stacked column chart for "Parental Level of Education" and "Test Preparation Course"
# First, get a cross-tabulation of the data
education_prep_crosstab = pd.crosstab(df['Parental Level of Education'], df['Test Preparation Course'])

# Plot the stacked column chart
education_prep_crosstab.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'salmon'], edgecolor='black')

# Customize the chart
plt.title('Parental Level of Education vs Test Preparation Course', fontsize=16)
plt.xlabel('Parental Level of Education', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Test Preparation Course', fontsize=12, title_fontsize=14)
plt.tight_layout()

# Show the chart
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency, ttest_ind
from statsmodels.stats.weightstats import ztest

df = pd.read_csv('StudentsPerformance.csv')

# Chi-Square Test
contingency_table = pd.crosstab(df['Gender'], df['Test Preparation Course'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Test:")
print("Chi2 Value:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

```

```python
import pandas as pd
from scipy.stats import ttest_ind
from statsmodels.stats.weightstats import ztest

df = pd.read_csv('StudentsPerformance.csv')

# Extract scores
math_scores = df['Math Score']
reading_scores = df['Reading Score']
writing_scores = df['Writing Score']


# Perform pairwise z-tests
z_test_results = {}
z_stat_math_reading, p_value_z_math_reading = ztest(math_scores, reading_scores)
z_test_results['Math vs Reading'] = (z_stat_math_reading, p_value_z_math_reading)

z_stat_math_writing, p_value_z_math_writing = ztest(math_scores, writing_scores)
z_test_results['Math vs Writing'] = (z_stat_math_writing, p_value_z_math_writing)

z_stat_reading_writing, p_value_z_reading_writing = ztest(reading_scores, writing_scores)
z_test_results['Reading vs Writing'] = (z_stat_reading_writing, p_value_z_reading_writing)

print("\nZ-Test Results:")
for pair, (z_stat, p_value) in z_test_results.items():
    print(f"{pair}: z-statistic = {z_stat:.4f}, p-value = {p_value:.4f}")
```
